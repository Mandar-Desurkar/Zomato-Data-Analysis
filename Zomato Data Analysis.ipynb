```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Zomato data .csv')
df
```

```python
def handleRate(value):
    value= str(value).split('/')
    value=value[0]
    return  float(value)

df['rate']= df['rate'].apply(handleRate)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['rate'], bins=10, edgecolor='black', alpha=0.7)

plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Rate', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Frequency', fontsize=14, fontweight='bold', color='darkblue')
plt.title('Histogram of Rates', fontsize=16, fontweight='bold', color='darkred')
```

```python
c = df['approx_cost(for two people)']

plt.figure(figsize=(10, 6))
ax = sns.countplot(x=c, hue=c, palette='viridis', edgecolor='black', dodge=False)

plt.xlabel('Approximate Cost for Two People', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='darkblue')
plt.title('Distribution of Approximate Cost for Two People', fontsize=16, fontweight='bold', color='darkred')

plt.grid(True, linestyle='--', alpha=0.7)

# Customize x-ticks
plt.xticks(rotation=45, ha='right', fontsize=12)

# Annotate bars
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width / 2, y + height), ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='online_order', y='rate', data=df, hue='online_order', palette='coolwarm', fliersize=8, linewidth=1.5, dodge=False)
plt.xlabel('Online Order', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Rate', fontsize=14, fontweight='bold', color='darkblue')
plt.title('Rate Distribution by Online Order Status', fontsize=16, fontweight='bold', color='darkred')

plt.grid(True, linestyle='--', alpha=0.7, which='both', axis='y')

plt.xticks(fontsize=12, rotation=0, ha='center')

plt.yticks(fontsize=12)

medians = df.groupby('online_order')['rate'].median()
for i, median in enumerate(medians):
    plt.hlines(median, xmin=i-0.4, xmax=i+0.4, color='black', linestyle='--', linewidth=1.2)

plt.tight_layout()
plt.show()
```
